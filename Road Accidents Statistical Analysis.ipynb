```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=[15,7],)
print('Countplot for Region')
sns.countplot(df['region'], data = df, palette = 'hls')
plt.xticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize=[15,7],)
print('Pieplot for Region')
df['region'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.xticks(rotation = 0)
plt.show()
```

```python
q=pd.pivot_table(df,index='region',values='signalavoid2019')
q=q.sort_values(by='signalavoid2019',ascending=False)
p=q.plot(kind='bar',figsize=(15,5),title='Amounts by region (signal avoid 2019)', color="pink")
p.bar_label(p.containers[0], label_type='center')
plt.show()
```

```python
q=pd.pivot_table(df,index='region',values='asleep2019')
q=q.sort_values(by='asleep2019',ascending=False)
p=q.plot(kind='bar',figsize=(15,5),title='Amounts by region (asleep 2019)', color="violet")
p.bar_label(p.containers[0], label_type='center')
plt.show()
```

```python
q=pd.pivot_table(df,index='region',values='othercause2019')
q=q.sort_values(by='othercause2019',ascending=False)
p=q.plot(kind='bar',figsize=(15,5),title='Amounts by region (other cause 2019)', color="grey")
p.bar_label(p.containers[0], label_type='center')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (20, 20)
fig= df.groupby('region')[['alcintake2019','overspeed2019','overtaking2019', 'lanejumping2019','wrongside2019', 'signalavoid2019', 'asleep2019', 'othercause2019']].agg('sum')
fig=fig.sort_values(by='region',ascending=False)
fig.plot(kind='bar',title='Total Purchases by Year')
plt.show()
```

```python
q=pd.pivot_table(df,index='region',values='asleep2021')
q=q.sort_values(by='asleep2021',ascending=False)
p=q.plot(kind='bar',figsize=(15,5),title='Amounts by region (asleep 2021)', color="blue")
p.bar_label(p.containers[0], label_type='center')
plt.show()
```
